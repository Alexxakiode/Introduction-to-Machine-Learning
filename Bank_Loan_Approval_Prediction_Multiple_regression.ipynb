USING MULTIPLE REGRESSION MODEL

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
#Defining the features and the label
X = bank_details[['default','balance','housing','loan']]
y = bank_details ['Target']
```

```python
#Splitting and training the data for test and train purpose
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X,y, test_size=0.3, random_state=1)

#Defining the Model
Loan_approval_model = LinearRegression()
Loan_approval_model
```

```python
#Multiple regression
# with sklearn
Loan_approval_model = LinearRegression()
Loan_approval_model.fit(X, y)

print('Intercept: \n', Loan_approval_model.intercept_)
print('Coefficients: \n', Loan_approval_model.coef_)
```

```python
pred = Loan_approval_model.predict(X_test)
```

```python
Loan_approval_model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print("Model Coefficients:", Loan_approval_model.coef_)
print("Mean Absolute Error:", mean_absolute_error(y_test, pred))
print("Coefficient of Determination:", r2_score(y_test, pred))
```
